```python
import requests

# Make a GET request to the Wikidata API to retrieve data for a specific entity
entity_id = 'Q41112'
api_url = 'https://www.wikidata.org/w/api.php?action=wbgetentities&format=json&ids=' + entity_id
response = requests.get(api_url)

# Extract the data for the entity from the response
entity_data = response.json()['entities'][entity_id]

# Print the label and description of the entity
print(entity_data['labels']['en']['value'])
print(entity_data['descriptions']['en']['value'])

# Print the properties and values of the entity
for prop, value in entity_data['claims'].items():
    print(prop, value[0]['mainsnak']['datavalue']['value'])

```

```python
import requests


# Set the SPARQL query that you want to execute
query = '''
PREFIX target: <http://www.wikidata.org/entity/Q41112>

SELECT
  DISTINCT
  ?start_date
  ?trial ?trialLabel
  ?intervention ?interventionLabel
  ?sponsor ?sponsorLabel
WHERE {
  ?trial wdt:P31 wd:Q30612 ;
  wdt:P1050 / wdt:P279* target: .
  OPTIONAL {
    ?trial wdt:P580 ?starttime
    BIND(SUBSTR(STR(?starttime), 0, 11) AS ?start_date)
  }
  OPTIONAL { ?trial wdt:P4844 ?intervention }
  OPTIONAL { ?trial wdt:P859 ?sponsor }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en,da,de,es,fr,jp,nl,no,pl,ru,sv,zh". }
}
ORDER BY DESC(?starttime)
'''

# Use the 'sparql' action to execute the query
url = 'https://query.wikidata.org/sparql'
params = {
    'format': 'json',
    'query': query
}
response = requests.get(url, params=params).json()

# Print the results of the query
print(response['results']['bindings'])
```

```python
import SPARQLWrapper

sparql = SPARQLWrapper.SPARQLWrapper("https://query.wikidata.org/sparql")

query = '''
PREFIX target: <http://www.wikidata.org/entity/Q41112>

SELECT
  DISTINCT
  ?start_date
  ?trial ?trialLabel
  ?intervention ?interventionLabel
  ?sponsor ?sponsorLabel
WHERE {
  ?trial wdt:P31 wd:Q30612 ;
  wdt:P1050 / wdt:P279* target: .
  OPTIONAL {
    ?trial wdt:P580 ?starttime
    BIND(SUBSTR(STR(?starttime), 0, 11) AS ?start_date)
  }
  OPTIONAL { ?trial wdt:P4844 ?intervention }
  OPTIONAL { ?trial wdt:P859 ?sponsor }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en,da,de,es,fr,jp,nl,no,pl,ru,sv,zh". }
}
ORDER BY DESC(?starttime)
'''

# Set the query language to SPARQL
sparql.setQuery(query)

# Set the return format to JSON
sparql.setReturnFormat(SPARQLWrapper.JSON)

# Execute the query and store the results
results = sparql.query().convert()

# Print the results
# print(results)

# print(results['results']['bindings'])


for trial in results['results']['bindings']:
    for key, value in trial.items():
        print(key)
        print(value['value'])
```

```python
import matplotlib.pyplot as plt

start_dates = []

for trial in results['results']['bindings']:
    for key, value in trial.items():
        if key == 'start_date':
            start_dates.append(value['value'])

plt.bar(start_dates, start_dates)
plt.ylabel('number of trials')
plt.show()
```

```python
start_dates = []
year_2022 = []
year_2021 = []
year_2020 = []
year_2019 = []
year_2018 = []
older = []

for trial in results['results']['bindings']:
    for key, value in trial.items():
        if key == 'start_date':
            start_dates.append(value['value'])
for date in start_dates:
    if date.startswith('2022'):
        year_2022.append(date)
    elif date.startswith('2021'):
        year_2021.append(date)
    elif date.startswith('2020'):
        year_2020.append(date)
    elif date.startswith('2019'):
        year_2019.append(date)
    elif date.startswith('2018'):
        year_2018.append(date)

    else:
        older.append(date)

print(year_2022)
print(len(year_2022))
print(year_2021)
print(len(year_2021))
print(year_2020)
print(len(year_2020))
print(year_2019)
print(len(year_2019))
print(year_2018)
print(len(year_2018))
print(older)
print(len(older))
```
